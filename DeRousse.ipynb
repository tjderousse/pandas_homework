## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order

* Display a preview of the summary data frame



```python
big_spender_count_df = purchase_data_df.groupby("SN").count()["Price"].rename("Purchase Count")
big_spender_avgprice_df = purchase_data_df.groupby("SN").mean()["Price"].rename("Average Purchase Price")
big_spender_total_df = purchase_data_df.groupby("SN").sum()["Price"].rename("Total Purchase Value")

big_spender_demo_df = pd.DataFrame({"Purchase Count":big_spender_count_df,
                                   "Average Purchase Price":big_spender_avgprice_df,
                                   "Total Purchase Value":big_spender_total_df})
#big_spender_demo_df.head()
topbig_spender_df = big_spender_demo_df.sort_values("Total Purchase Value", ascending=False)
topbig_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_count_df = purchase_data_df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Number of Purchases")
avg_items_price_df = purchase_data_df.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Average Purchase Price")
total_items_value_df = purchase_data_df.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Value of Purchase")
topitems_df = pd.DataFrame({"Number of Purchases":items_count_df,
                            "Average Purchase Price":avg_items_price_df,
                            "Total Value of Purchase":total_items_value_df})
#topitems_df
```

```python
topitems_sorted_df = topitems_df.sort_values("Number of Purchases", ascending=False)
topitems_sorted_df.head()
```
